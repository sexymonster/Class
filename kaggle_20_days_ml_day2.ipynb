나는 데이터를 구글 드라이브에 저장했기 때문에 아래처럼 쉽게 로드할 수 있다.

데이터 프레임이 어떻게 생겼는지 확인했으면 이제는 데이터의 요약을 확인해볼 차례다.

```python
melbourne_data.describe()
```

### 데이터 요약 해석

- count : 결측치가 아닌 데이터의 수를 나타낸다.
- mean : 값들의 평균을 나타낸다.
- std : 숫자가 얼마나 분산 되어있는지를 보여주는 표준편차다.
- min, 25%, 50%, 75%, max : 0부터 100까지 백분위수에 해당하는 값을 나타낸다.
